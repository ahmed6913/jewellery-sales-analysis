```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("sales_data.csv")
df
```

```python
df.info()
```

```python
df["Revenue"] = df["Price"] * df["Quantity"]
df
```

```python
df["Revenue"].sum()
```

```python
df.groupby("Product")["Revenue"].sum().plot(kind="bar")

plt.title("Revenue by Product")
plt.show()
```
